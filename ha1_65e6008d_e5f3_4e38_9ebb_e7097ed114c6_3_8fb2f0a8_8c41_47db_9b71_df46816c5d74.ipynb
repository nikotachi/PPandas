8. Чтобы посчитать итоговую зарплату, нам нужно по каждому человеку знать и оклад, и премию. Для этого надо будет соединить (сджойнить) таблицы по `person_id`.
```python
df1 = df.merge(bonus, on="person_id", how="outer")
df1
```
