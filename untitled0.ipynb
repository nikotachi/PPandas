```python
services = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
            'TechSupport', 'StreamingTV', 'StreamingMovies']

# Подсчет числа подключенных услуг для каждого клиента
df["ActiveServices"] = df[services].apply(lambda x: (x == 'Yes').sum(), axis=1)

# Фильтрация клиентов с 4 услугами
df4 = df[df['ActiveServices'] == 4]

# Разделение на Fiber optic и DSL
fiber = df4[df4['InternetService'] == 'Fiber optic']
dsl = df4[df4['InternetService'] == 'DSL']

# Подсчет оттока в каждой группе
churn_fiber = fiber['Churn'].value_counts(normalize=True).get('Yes', 0) * 100
churn_dsl = dsl['Churn'].value_counts(normalize=True).get('Yes', 0) * 100

# Разница в оттоке
difference = churn_fiber - churn_dsl

print(f"Отток у Fiber optic: {churn_fiber:.2f}%")
print(f"Отток у DSL: {churn_dsl:.2f}%")
print(f"Разница: {difference:.2f}% (Fiber optic выше на {abs(difference):.2f}%)")
```
