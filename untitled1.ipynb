№5 Посчитаем количество заказов на каждом цехе с помощью value_counts(), затем используем idxmax() для нахождения названия цеха с максимальным количеством заказов. Получим "Цех 3"

№7 Сгрупируем с помощью функции groupby по столбцу деталей, при этом второй столбец "Время изготовления (мин)" возьмем среднее значения с помощью функции mean() и для того чтобы преобразовать в удобный табличный формат используем reset_index()

```python
df.groupby('Тип детали')['Время изготовления (мин)'].mean().reset_index()
```

№9 Сгруппируем таблицу по "Количеству" и посчитаем для каждого количества шанс брака, используя функцию lambda по ключу "Брак". Добавим столбец "Шанс брака". Получим следующую табличку

```python
df7 = df.groupby("Количество")["Качество"].apply(lambda x : (x == "Брак").mean()*100).reset_index(name = "Шанс брака")
df7
```

Попробуем визуализировать с помощью линейной диаграммы, она лучше всего отобразит изменение шанса брака от количества заказов.

```python
from matplotlib import pyplot as plt
df7['Шанс брака'].plot(kind='line', figsize=(8, 4), title='Шанс брака')
```

График показывает значительные колебания шанса брака на всём диапазоне заказов, без очевидного тренда — он не становится заметно выше или ниже с увеличением количества заказов. Значит шанс брака не зависит от количества заказов.

№10 Выведем топ 5 самых популярных типов деталей используя группировку по типу детали и суммируем их количество. Далее отсортируем по убыванию количества деталей. Получим следующую таблицу

```python
df8 = df.groupby("Тип детали")["Количество"].sum().reset_index()
df8.sort_values(by='Количество', ascending=False)
```
